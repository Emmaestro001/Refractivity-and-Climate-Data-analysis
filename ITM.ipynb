```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some example data
x = np.linspace(0, 30, 90)  # 90 points
y = np.sin(x) + np.random.normal(scale=0.1, size=len(x))

# Split the data into thirds
n = len(x)
third = n // 3

# Plot each third with a different color
plt.scatter(x[:third], y[:third], color='red', label='First third')
plt.scatter(x[third:2*third], y[third:2*third], color='green', label='Second third')
plt.scatter(x[2*third:], y[2*third:], color='blue', label='Final third')

plt.legend()
plt.title('Scatter Plot with Data in Thirds')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
