```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
REF= pd.read_excel('CORRELATION DATA.xlsx', sheet_name ='REF')
TEMP= pd.read_excel('CORRELATION DATA.xlsx', sheet_name ='TEMP')
PRES= pd.read_excel('CORRELATION DATA.xlsx', sheet_name ='PRES')
RH= pd.read_excel('CORRELATION DATA.xlsx', sheet_name ='RH')
PRECIP= pd.read_excel('CORRELATION DATA.xlsx', sheet_name ='PRECIP')
```

```python
import matplotlib.gridspec as gridspec

# create a figure with a grid of 3 rows and 2 columns
fig = plt.figure(figsize=(10, 15)) 
gs = gridspec.GridSpec(nrows=3, ncols=2, figure=fig, hspace=0.3, wspace=0.00)

# create Seaborn plots for each axis
ax1 = fig.add_subplot(gs[0, 0])
sns.boxplot(data=TEMP, orient='h', ax=ax1)
plt.setp(ax1.get_xticklabels(), rotation=90)
ax1.set_xlabel('Temperature ($^oC$)')
ax1.set_title('TEMPERATURE', fontweight='bold')


ax2 = fig.add_subplot(gs[0, 1])
sns.boxplot(data=RH, orient='h', ax=ax2)
plt.setp(ax2.get_xticklabels(), rotation=90)
ax2.set_xlabel('Relative Humidity (%)')
ax2.set_title('RELATIVE HUMIDITY',  fontweight='bold')
ax2.set(yticklabels=[], ylabel='')

ax3 = fig.add_subplot(gs[1, 0])
sns.boxplot(data=PRES, orient='h', ax=ax3)
plt.setp(ax3.get_xticklabels(), rotation=90)
ax3.set_xlabel('Atmospheric Pressure (hPa)')
ax3.set_title('ATMOSPHERIC PRESSURE', fontweight='bold')

ax4 = fig.add_subplot(gs[1, 1])
sns.boxplot(data=PRECIP, orient='h', ax=ax4)
plt.setp(ax4.get_xticklabels(), rotation=90)
ax4.set_xlabel('Precipitation (mm)')
ax4.set_title('PRECIPITATION',  fontweight='bold')
ax4.set(yticklabels=[], ylabel='')

ax5 = fig.add_subplot(gs[2, :])
sns.boxplot(data=REF, ax=ax5)
ax5.set_title('REFRACTIVITY',  fontweight='bold')
ax5.set_ylabel('Refractivity (N-Units)')


# display the plot
plt.show()
```
