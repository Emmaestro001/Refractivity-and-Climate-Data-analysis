```python
import pandas as pd
import matplotlib.pyplot as plt
import pymannkendall as mk
import seaborn as sns
%matplotlib inline
```

```python
#IMPORT EXCEL SHEETS CONTAINING DATA FOR EACH LOCATION

YEN= pd.read_excel('CORRELATION DATA.xlsx', sheet_name ='YEN', index_col='YEAR') #Yenagoa
IK= pd.read_excel('CORRELATION DATA.xlsx', sheet_name ='IK', index_col='YEAR') #Ikot Akpaden
IBN= pd.read_excel('CORRELATION DATA.xlsx', sheet_name ='IBN', index_col='YEAR') #Ibadan
ABJ= pd.read_excel('CORRELATION DATA.xlsx', sheet_name ='ABJ', index_col='YEAR') #Abuja
JAL= pd.read_excel('CORRELATION DATA.xlsx', sheet_name ='JAL', index_col='YEAR') #Jalingo
SOK= pd.read_excel('CORRELATION DATA.xlsx', sheet_name ='SOK', index_col='YEAR') #Sokoto
GOM= pd.read_excel('CORRELATION DATA.xlsx', sheet_name ='GOM', index_col='YEAR') #Gombe
KUK= pd.read_excel('CORRELATION DATA.xlsx', sheet_name ='KUK', index_col='YEAR') #Kukawa

#WE CAN VIEW THE FIRST FEW ROWS OF THE FIRST TABLE
YEN.head()
```

```python
tablea=YEN.describe()
tableb=IK.describe()
tablec=IBN.describe()
tabled=ABJ.describe()
tablee=JAL.describe()
tablef=SOK.describe()
tableg=GOM.describe()
tableh=KUK.describe()

tableh
```

```python
tablea.to_excel('YEN.xlsx')
tableb.to_excel('IK.xlsx')
tablec.to_excel('IBN.xlsx')
tabled.to_excel('ANJ.xlsx')
tablee.to_excel('JAL.xlsx')
tablef.to_excel('SOK.xlsx')
tableg.to_excel('GOM.xlsx')
tableh.to_excel('KUK.xlsx')
```

```python
#FIND THE CORRELATION FOR EACH TABLE

YEN_CORR = YEN.corr() #Yenagoa
IK_CORR = IK.corr() #Ikot Akpaden 
IBN_CORR = IBN.corr() #Ibadan
ABJ_CORR = ABJ.corr() #Abuja
JAL_CORR = JAL.corr() #Jalingo
SOK_CORR = SOK.corr() #Sokoto
GOM_CORR = GOM.corr() #Gombe
KUK_CORR = KUK.corr() #Kukawa

#We can view the correlation results for Yenagoa
YEN_CORR
```

```python
# Put the correlation matrices and their titles in a list for easy iteration
corr_matrices = [
    (YEN_CORR, "Yenagoa"),
    (IK_CORR, "Ikot Akpaden"),
    (IBN_CORR, "Ibadan"),
    (ABJ_CORR, "Abuja"),
    (JAL_CORR, "Jalingo"),
    (SOK_CORR, "Sokoto"),
    (GOM_CORR, "Gombe"),
    (KUK_CORR, "Kukawa")
]

# Set up the matplotlib figure (4 rows x 2 columns)
fig, axes = plt.subplots(nrows=4, ncols=2, figsize=(10, 20))
fig.suptitle("Correlation Heatmaps of Meteorological Parameters", fontsize=20)

# Flatten the axes array for easy iteration
axes = axes.flatten()

# Set a color map and color scale limits for uniformity
#cmap = sns.diverging_palette(260, 15, as_cmap=True)
#cmap = 'RdBu_r'
cmap = 'coolwarm'
vmin = -1
vmax = 1

# Plot each heatmap
for i, (corr, title) in enumerate(corr_matrices):
    sns.heatmap(corr, ax=axes[i], cmap=cmap, vmin=vmin, vmax=vmax, cbar=(i==0),
                cbar_ax=None if i else fig.add_axes([.91, .3, .03, .4]), annot=True, fmt=".2f")
    axes[i].set_title(title)
    axes[i].tick_params(axis='x', rotation=45)
    axes[i].tick_params(axis='y', rotation=0)

# Adjust layout
plt.tight_layout(rect=[0, 0, 0.9, 0.95])  # Leave space for colorbar and title
plt.show()
```
